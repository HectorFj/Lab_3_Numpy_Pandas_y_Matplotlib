1) ¿Cuántos Municipios tienen más de 100000

```python
# Se selecciona los municipios que tienen mas de 100k habitantes
municipios = dataset[dataset['Poblacion'] > 100000]
#print(municipios_con_100k)

# Contar cuantos municipios son
cantidad_municipios = len(municipios)

print("El número de municipios con más de 100k de habitantes es:", cantidad_municipios)
```

2) Realice una gráfica de barras sobre la polbación de cada 
ciudad, ordenela de menor a mayo y responda:
a. ¿Cuál es la segunda ciudad más poblada?
b. ¿Qué posición ocupa Granada en el ranking de 
las ciudades más pobladas?

```python
total_poblacion = dataset.groupby('Provincia')['Poblacion'].sum().reset_index()

poblacion_por_provincia = total_poblacion.sort_values('Poblacion', ascending=False)

plt.figure(figsize=(12, 10))
plt.bar(poblacion_por_provincia['Provincia'], poblacion_por_provincia['Poblacion'])
plt.xlabel('Ciudad')
plt.ylabel('Población')
plt.title('Población de cada ciudad')
plt.xticks(rotation=90)
plt.show()

segunda_ciudad_mas_poblada = dataset.groupby('Municipio')['Poblacion'].sum().nlargest(2).index[1]
print("El nombre de la segunda ciudad más poblada es:", segunda_ciudad_mas_poblada)
```

b. ¿Qué posición ocupa Granada en el ranking de 
las ciudades más pobladas?

```python
posicion_granada = poblacion_por_provincia[poblacion_por_provincia['Provincia'] == 'Granada'].index[0] + 0
print("Granada ocupa la posición número:", posicion_granada)
```

5) Obtenga la media, mediana, desviación estándar, valor 
máximo y valor mínimo de la población de los municipios 
de la provincia de Granada.

```python
municipios_granada = dataset[dataset['Provincia'] == 'Granada']

media = municipios_granada['Poblacion'].mean()
mediana = municipios_granada['Poblacion'].median()
desviacion = municipios_granada['Poblacion'].std()
valor_maximo = municipios_granada['Poblacion'].max()
valor_minimo = municipios_granada['Poblacion'].min()

print("Media de la población:", media)
print("Mediana de la población:", mediana)
print("Desviación estándar de la población:", desviacion)
print("Valor máximo de la población:", valor_maximo)
print("Valor mínimo de la población:", valor_minimo)
```

6) Realice un histograma con la población de los 
Municipios para cada una de las provincias.

```python
provincias = dataset['Municipio'].unique()
# Provincia
# Municipio
plt.figure(figsize=(20, 10))
for provincia in provincias:
    poblacion_provincia = dataset[dataset['Provincia'] == provincia]['Poblacion']
    plt.hist(poblacion_provincia, bins=20, alpha=0.7, label=provincia)

plt.xlabel('Población')
plt.ylabel('Frecuencia')
plt.title('Histograma de población por provincia')
plt.legend(title='Provincia', loc='upper right')
plt.show()
```
